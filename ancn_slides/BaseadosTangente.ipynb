# Métodos baseados em tangente (em construção)

```python
def Newton(f, df, x, tol, maxiter=10):
    for i in range(maxiter):
        DeltaX = f(x)/df(x)
        x = x - DeltaX
        print(i, x, f(x))
        if abs(DeltaX) < tol or abs(f(x)) < tol:
            break
    return x
```

```python
import numpy as np

def f(x):
    return 2*x**3 - np.cos(x+1) -3

def df(x):
    return 6*x**2 + np.sin(x+1)

Newton(f, df, 0.5, 1e-2)
```
